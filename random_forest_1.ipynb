```python
%load_ext autoreload
%autoreload 2

from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as plt
from spectrum import Spectrum
```

```python
hd185859 = Spectrum('HD185859')
```

```python
X = hd185859.wavelength.reshape(-1, 1)
y = hd185859.flux

print(X)

rf = RandomForestRegressor(n_estimators=500, random_state=42)
rf.fit(X, y)

# Make predictions
y_rf = rf.predict(X)
```

```python
print(y_rf)
margin = 2
center = 6630

fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(15, 9), sharex=True)

axes[0].plot(hd185859.wavelength, hd185859.flux, '.')
axes[1].plot(hd185859.wavelength, y_rf, '.')
axes[2].plot(hd185859.wavelength, hd185859.flux - y_rf, '.')
axes[0].set_xlim(center - margin, center + margin)
```
