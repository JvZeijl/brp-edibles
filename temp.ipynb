```python
import numpy as np
import matplotlib.pyplot as plt

file_pattern = 'data/ascii/HD170740/HD170740_w564_n9_20160612_U.txt'
na_line1 = 5889.95
na_line2 = 5895.924

wavelength, flux = np.loadtxt(file_pattern, unpack=True)

fig, axes = plt.subplots(2, 2, figsize=(15, 3))

axes[0][0].plot(wavelength, flux)
axes[0][0].axvline(na_line1, color='orange')
axes[0][0].axvline(na_line2, color='red')

axes[0][0].set_xlabel('Wavelength')
axes[0][0].set_ylabel('Intensity')
axes[0][0].set_xlim(na_line1 - 10, na_line2 + 10)

fig.tight_layout()
plt.show()
```
